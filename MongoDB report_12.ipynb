### 2. Data Overview

The **movies** collection, sourced from the Mflix dataset, holds **23,149 documents**, occupying approximately **1.2 GB** of storage within MongoDB Atlas.  
Each document represents a movie and contains attributes such as title, genre, cast, and ratings.  
Being semi-structured (BSON format), the data contains nested fields and arrays, allowing for complex and flexible queries.

Each document generally includes:

- **_id (ObjectId):**  
  Unique identifier for each movie.  
  Example: `"5a934e000102030405000000"`

- **title (String):**  
  Movie name.  
  Example: `"Inception"`

- **year (Number):**  
  Year the movie was released.  
  Example: `2010`

- **genres (Array of Strings):**  
  Categories the movie belongs to.  
  Example: `["Action", "Sci-Fi", "Thriller"]`

- **cast (Array of Strings):**  
  List of actors featured.  
  Example: `["Leonardo DiCaprio", "Joseph Gordon-Levitt", "Elliot Page"]`

- **directors (Array of Strings):**  
  Directors' names.  
  Example: `["Christopher Nolan"]`

- **writers (Array of Strings):**  
  Writers or screenwriters.  
  Example: `["Christopher Nolan", "Jonathan Nolan"]`

- **languages (Array of Strings):**  
  Languages the movie is available in.  
  Example: `["English", "Japanese", "French"]`

- **countries (Array of Strings):**  
  Countries where the movie was produced.  
  Example: `["USA", "UK"]`

- **released (Date):**  
  Official release date.  
  Example: `ISODate("2010-07-16T00:00:00Z")`

- **runtime (Number):**  
  Movie duration in minutes.  
  Example: `148`

- **plot (String):**  
  A brief storyline summary.  
  Example:  
  `"A thief who steals corporate secrets through dream-sharing technology is given a chance to erase his criminal record."`

- **fullplot (String):**  
  A detailed plot summary.

- **imdb (Object):**  
  IMDb specific details:  
  - **rating (Number):** IMDb rating (e.g., 8.8)  
  - **votes (Number):** Number of votes received (e.g., 2,000,000)  
  - **id (Number):** IMDb movie ID (e.g., 1375666)

- **tomatoes (Object):**  
  Rotten Tomatoes related data:  
  - **viewer (Object):** Viewer rating and reviews.  
  - **rating (Number):** Viewer rating (e.g., 4.2)  
  - **numReviews (Number):** Viewer reviews count (e.g., 5000)  
  - **critic (Object):** Critic reviews.  
  - **fresh (Number):** Fresh ratings count.  
  - **rotten (Number):** Rotten ratings count.  
  - **lastUpdated (Date):** Last update timestamp.

- **type (String):**  
  Type of media (generally `"movie"`).

### 7. Observations and Findings

**1. Movie Distribution by Genre**
- Genres like **Drama**, **Action**, and **Comedy** dominate movie production, making them the most widely created genres.
- **Sci-Fi** and **Thriller** films often secure higher ratings, indicating a smaller but highly engaged fan base.

**2. IMDb Ratings Analysis**
- The majority of films have IMDb ratings between **6.0 and 8.0**, with very few exceeding **8.5**.
- High-rated movies are predominantly in the **Drama** and **Thriller** categories, showing a preference for well-constructed narratives.
- Movies rated above **8.0** usually maintain longer audience engagement times.

**3. Trends in Yearly Movie Production**
- Movie production surged significantly after **2000**, highlighting the influence of **digital media and streaming services**.
- During the **1980s and 1990s**, movie output remained steady, but a sharp growth trajectory started in the new millennium.

**4. Language Distribution**
- **English** remains the leading language in the dataset, followed by noticeable contributions from **Spanish**, **French**, and **Hindi**.
- A growing trend of **multilingual films** reflects the film industry's increasing globalization.

**5. Runtime Analysis**
- Most movies have runtimes between **90 to 150 minutes**, with only a few exceeding the **3-hour mark**.
- **Short films** and **documentaries** generally have much shorter durations.

**6. Awards and Ratings Correlation**
- Films with higher IMDb ratings are more likely to be **nominated for or win awards**.
- Thereâ€™s a strong link between **higher Rotten Tomatoes critic scores** and **award wins**.

- A notable **gap often exists** between **critic (Metacritic)** and **audience (IMDb)** ratings, suggesting differing perspectives on movie quality.

- Some niche genres like **Film Noir** showcase extremely high Metacritic ratings, but due to a **limited number of films**, these averages can appear inflated.

- Emphasize promoting **award-winning** and **critically acclaimed** films to niche audiences.
